```python
bagging = BaggingClassifier(base_estimator=decision_tree, n_estimators=5, random_state=42)
bagging.fit(X_train, y_train)
bagging_pred = bagging.predict(X_test)
bagging_accuracy = accuracy_score(y_test, bagging_pred)
bagging_precision = precision_score(y_test, bagging_pred)
bagging_recall = recall_score(y_test, bagging_pred)
bagging_cm = confusion_matrix(y_test, bagging_pred)
```

```python
# Boosting with Decision Tree Classifier (AdaBoost)
boosting = AdaBoostClassifier(base_estimator=decision_tree, n_estimators=5, random_state=42)
boosting.fit(X_train, y_train)
boosting_pred = boosting.predict(X_test)
boosting_accuracy = accuracy_score(y_test, boosting_pred)
boosting_precision = precision_score(y_test, boosting_pred)
boosting_recall = recall_score(y_test, boosting_pred)
boosting_cm = confusion_matrix(y_test, boosting_pred)
```

```python
random_forest = RandomForestClassifier(n_estimators=5, random_state=42)
random_forest.fit(X_train, y_train)
rf_pred = random_forest.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)
rf_precision = precision_score(y_test, rf_pred)
rf_recall = recall_score(y_test, rf_pred)
rf_cm = confusion_matrix(y_test, rf_pred)
```

```python
print("Bagging Accuracy:", bagging_accuracy)
print("Bagging Precision:", bagging_precision)
print("Bagging Recall:", bagging_recall)
print("Bagging Confusion Matrix:\n", bagging_cm)
print("Boosting Accuracy:", boosting_accuracy)
print("Boosting Precision:", boosting_precision)
print("Boosting Recall:", boosting_recall)
print("Boosting Confusion Matrix:\n", boosting_cm)
print("Random Forest Accuracy:", rf_accuracy)
print("Random Forest Precision:", rf_precision)
print("Random Forest Recall:", rf_recall)
print("Random Forest Confusion Matrix:\n", rf_cm)
```
